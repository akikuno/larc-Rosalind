```python
spectrums = [
    3524.8542,
3710.9335,
3841.974,
3970.0326,
4057.0646,
]
```

```python
MASS: dict[float, str] = {
    71.03711: "A",
    103.00919: "C",
    115.02694: "D",
    129.04259: "E",
    147.06841: "F",
    57.02146: "G",
    137.05891: "H",
    113.08406: "I",  # Isoleucine or Leucine
    128.09496: "K",
    131.04049: "M",
    114.04293: "N",
    97.05276: "P",
    128.05858: "Q",
    156.10111: "R",
    87.03203: "S",
    101.04768: "T",
    99.06841: "V",
    186.07931: "W",
    163.06333: "Y",
}
```

```python
MASS_VALUES = sorted(MASS.keys())

from bisect import bisect_left, bisect_right

ans = []
for s1, s2 in zip(spectrums, spectrums[1:]):
    diff = s2 - s1 - 0.01
    i = bisect_left(MASS_VALUES, diff)
    ans.append(MASS[MASS_VALUES[i]])

print("".join(ans))    
```

```python
from pathlib import Path
data = Path("rosalind_spec.txt").read_text().strip().splitlines()
spectrums = [float(x) for x in data]

ans = []
for s1, s2 in zip(spectrums, spectrums[1:]):
    diff = s2 - s1 - 0.01
    i = bisect_left(MASS_VALUES, diff)
    ans.append(MASS[MASS_VALUES[i]])

print("".join(ans))    
```
